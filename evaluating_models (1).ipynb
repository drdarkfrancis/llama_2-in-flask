```python
import PyPDF2

def extract_text_from_pdf(pdf_file):
    with open(pdf_file, 'rb') as file:
        reader = PyPDF2.PdfReader(file)
        text = ""
        for page in reader.pages:
            text += page.extract_text()
    return text
```

# Defining the models



1.   BERT legal for question answering
2.   roBERTa finetuned on CUAD for QA



```python
import torch
from transformers import BertForQuestionAnswering, BertTokenizer

# Load the pre-trained BERT model and tokenizer
model_name = "atharvamundada99/bert-large-question-answering-finetuned-legal"
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForQuestionAnswering.from_pretrained(model_name)
```

```python
def get_answer_bert(question, context):
    # Encode the inputs
    inputs = tokenizer.encode_plus(question, context, add_special_tokens=True, return_tensors="pt")

    # Check if the input length is within the limit
    input_ids = inputs["input_ids"].tolist()[0]
    if len(input_ids) > 512:
        return "Error: The context is too long.", len(input_ids)

    # Model input values
    input_segment_ids = inputs["token_type_ids"]
    input_mask = inputs["attention_mask"]

    # Predict answer tokens
    with torch.no_grad():
        outputs = model(input_ids=inputs["input_ids"], token_type_ids=input_segment_ids, attention_mask=input_mask)
        answer_start_scores, answer_end_scores = outputs.start_logits, outputs.end_logits

    # Debugging: Inspect the model's predictions
    answer_start = torch.argmax(answer_start_scores)
    answer_end = torch.argmax(answer_end_scores) + 1
    print("Debug: Start position:", answer_start.item(), "End position:", answer_end.item())

    # Convert tokens to answer string
    answer = tokenizer.convert_tokens_to_string(tokenizer.convert_ids_to_tokens(input_ids[answer_start:answer_end]))

    return answer
```

```python
from transformers import pipeline

# Load the question-answering pipeline with the specified model
qa_pipeline = pipeline("question-answering", model="akdeniz27/roberta-large-cuad")
```

```python
def ask_roberta(question, context):
    # Use the pipeline to get the answer
    result = qa_pipeline(question=question, context=context)
    return result['answer']
```

```python
def ask_llama(question,context):
  system_prompt = "you are a professional legal contractor ,answer the question based on the provided context"
  user_message=f"context:{context} , the question is : {question}"

  prompt = f"<s>[INST] <<SYS>>\n{system_prompt}\n<</SYS>>\n\n{user_message} [/INST]"
  answer=llama(prompt)
  return answer


```

```python
def ask_llama(question, context):
    system_prompt = "you are a german legal contractor, answer the question based on the provided context."

    # Calculate the maximum length allowed for context and question
    max_length = 512 - len(system_prompt) - len("<s>[INST] <<SYS>>\n\n<</SYS>>\n\n[/INST]")

    # Truncate context and question to fit within max_length
    user_message = f"context:{context} , the question is: {question}"
    if len(user_message) > max_length:
        # Truncate with preference to the question
        available_space = max_length - len(f" , the question is: {question}")
        truncated_context = context[:available_space]
        user_message = f"context:{truncated_context} , the question is: {question}"

    prompt = f"<s>[INST] <<SYS>>\n{system_prompt}\n<</SYS>>\n\n{user_message} [/INST]"
    answer = llama(prompt)
    return answer
```

# Testing the models

```python
def test_model(df, get_answer, db):
    for index, row in df.iterrows():
        # retrieve relevant data with the similarity search
        query = row['Questions']
        relevant_data = db.similarity_search_with_score(query, k=10)

        if relevant_data and len(relevant_data) >= 2:
            extracted_passage_1 = relevant_data[0][0].page_content  # first relevant passage
            extracted_passage_2 = relevant_data[1][0].page_content  # second relevant passage
            # update the DataFrame
            df.at[index, 'Extracted_passage_1'] = extracted_passage_1
            df.at[index, 'score_1'] = relevant_data[0][1]
            df.at[index, 'Extracted_passage_2'] = extracted_passage_2
            df.at[index, 'score_2'] =relevant_data[1][1]

            if relevant_data[0][1]>1.3:
              df.at[index,'Is_existant']=0
            else:
              df.at[index,'Is_existant']=1
              #context = extracted_passage_1 + " " + extracted_passage_2
              context = extracted_passage_1
              # Get the answer from the model
              model_answer = get_answer(query, context)
              df.at[index, 'QA_response'] = model_answer

    return df
```
